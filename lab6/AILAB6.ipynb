```python
def enumerateNeighbourStates(s):
    # Generates all possible neighbour states for a given state 's'.
    neighbours = []
    for col in range(8):  # Iterate through each column
        for row in range(1, 9):  # Try moving the queen in the current column to a different row
            # Skip the current row of the queen, as no change is needed if the queen stays in the same row
            if s[col] != row:
                new_state = s[:]  # Create a copy of the state to avoid modifying the original state
                new_state[col] = row  # Update the row of the queen in the current column to the new row
                neighbours.append(new_state)  # Add the new state to the list of neighbours
    return neighbours

def objectivefunctionvalue(s):
    # Calculates the heuristic value for a given state 's'.
    # Heuristic value = number of queen pairs that are attacking each other through 1) rows and 2) diagonals
    count = 0
    n = len(s)
    # Nested loop for checking attacks
    for i in range(8):
        for j in range(i + 1, 8):
            # Check if queens are in the same row
            if s[i] == s[j]:
                count += 1
            # Check if queens are on the same diagonal
            if abs(i - j) == abs(s[i] - s[j]):
                count += 1
    return -count  # Return negative count as we minimize attacking pairs

def hillClimbing(initial_state):
    current_state = initial_state
    current_value = objectivefunctionvalue(current_state)
    print("Initial State:", current_state)
    print("Initial Objective Function Value:", current_value)

    while True:
        # Generate all neighbor states and their heuristic values
        neighbours = enumerateNeighbourStates(current_state)
        neighbour_values = [objectivefunctionvalue(state) for state in neighbours]

        # Find the best neighbor (with the highest heuristic value)
        best_value = max(neighbour_values)
        best_state = neighbours[neighbour_values.index(best_value)]

        # If the best neighbor is not better than the current state, stop
        if best_value <= current_value:
            break

        # Otherwise, move to the best neighbor
        current_state = best_state
        current_value = best_value

        print(f"Moved to new state: {current_state} with objective function value: {current_value}")

    return current_state, current_value

# Test with the initial state
# Sn = [1, 3, 5, 7, 2, 4, 6, 8]
Sn= [1, 8, 1, 1, 1, 1, 1, 7]
final_state, final_value = hillClimbing(Sn)

print("Final State:", final_state)
print("Final Objective Function Value:", final_value)
```

```python
import random
import math

def enumerateNeighbourStates(s):
    neighbours = []
    for col in range(8):  # Iterate through each column
        for row in range(1, 9):  # Try moving the queen in the current column to a different row
            if s[col] != row:
                new_state = s[:]  # Create a copy of the state
                new_state[col] = row  # Update the row of the queen in the current column to the new row
                neighbours.append(new_state)  # Add the new state to the list of neighbours
    return neighbours

def objectivefunctionvalue(s):
    count = 0
    n = len(s)
    for i in range(8):
        for j in range(i + 1, 8):
            # Check if queens are in the same row
            if s[i] == s[j]:
                count += 1
            # Check if queens are on the same diagonal
            if abs(i - j) == abs(s[i] - s[j]):
                count += 1
    return -count  # Return negative count as we minimize attacking pairs

def simulatedAnnealing(initial_state, sch):
    current_state = initial_state
    current_value = objectivefunctionvalue(current_state)
    print("Initial State:", current_state)
    print("Initial Objective Function Value:", current_value)

    t = 0  # Start at the first time step
    while t < len(sch):  # Until the schedule runs out
        T = sch[t]  # Get the current temperature

        if T == 0:  # If temperature is zero, return the current state
            break

        # Generate a random neighbor
        neighbours = enumerateNeighbourStates(current_state)
        neighbor = random.choice(neighbours)
        neighbour_value = objectivefunctionvalue(neighbor)

        # Calculate the energy difference
        delta_E = neighbour_value - current_value

        # If the neighbor is better, move to it
        if delta_E > 0:
            current_state = neighbor
            current_value = neighbour_value
        # If the neighbor is worse, move to it with some probability
        else:
            probability = math.exp(delta_E / T)
            if random.random() < probability:
                current_state = neighbor
                current_value = neighbour_value

        # Increment time step
        t += 1

        print(f"Temperature: {T}, Current State: {current_state}, Objective Value: {current_value}")

    return current_state, current_value

# Example usage of simulated annealing with a temperature schedule
# Example schedule: Start at 1000 temperature and decrease by a factor of 0.95 each time
schedule = [1000 * (0.95 ** t) for t in range(100)]  # Decreasing temperature schedule
Sn = [5, 3, 5, 7, 2, 4, 6, 8]  # Initial state

final_state, final_value = simulatedAnnealing(Sn, schedule)

print("Final State:", final_state)
print("Final Objective Function Value:", final_value)
```
