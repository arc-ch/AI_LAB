```python
# 1) make 8 -> N
# 2) Stop when 1st soln Found
# 3) add time fn
# 4) do for n = 5 to 20 and plot table of n and time

import time

N = 23 # Set N

# Fitness function: calculate the number of conflicts
def f(s):  # Objective function
    count = 0
    for i in range(N):
        if s[i] is None: continue
        for j in range(i + 1, N):
            if s[j] is None: continue
            if s[i] == s[j]:
                count += 1
            if abs(i - j) == abs(s[i] - s[j]):
                count += 1
    return -count

# Backtracking function to solve the N-queens problem
def solve(s, col):

    if col == N:
        if f(s) == 0:  # No conflicts, valid solution
            print(s, f(s))
            return True  # Stop after finding the first solution

    for i in range(N):
        t = s.copy()  # Copy curr state
        t[col] = i    # Place the queen in the row i of the current column col

        # Check if the placement is valid
        if f(t) == 0:  # No queens attacking each other
            if solve(t, col + 1):  # Recurse to next col, stop when solution is found
                return True  # Stop after finding the first solution

    return False  # Return False if no solution found

# Start timing before calling solve
start_time = time.time()

solve([None] * N, 0)

# End timing after solve finishes
end_time = time.time()

# Calculate and print the elapsed time
execution_time = end_time - start_time
print(f"Execution Time: {execution_time:} seconds")
```
