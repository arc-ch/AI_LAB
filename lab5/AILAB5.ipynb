```python
# Implement A* Algorithm

# 1) adding heuristic value in city
# 2) changing priority function to g(n) + h(n)


# Dict of City with name and heuristic value
heuristics = {
    "Arad": 366,
    "Bucharest": 0,
    "Fagaras": 176,
    "Oradea": 380,
    "Pitesti": 100,
    "Rimnicu-Vilcea": 193,
    "Sibiu": 253,
}
# print(type(city))

# Adjacency list => Dict of Dicts
road = {
    "Arad": {"Sibiu": 140},                     # Arad to Sibiu=> distance = 140
    "Bucharest": {"Fagaras": 211, "Pitesti": 101},
    "Fagaras": {"Bucharest": 211, "Sibiu": 99},
    "Oradea": {"Sibiu": 151},
    "Pitesti": {"Bucharest": 101, "Rimnicu-Vilcea": 97},
    "Rimnicu-Vilcea": {"Pitesti": 97, "Sibiu": 80},
    "Sibiu": {"Arad": 140, "Fagaras": 99, "Oradea": 151, "Rimnicu-Vilcea": 80},
}

# Start and goal
init = "Arad"
goal = "Bucharest"  # Goal city

d = {}
q = []
world = {"id": -1}

# func to return unique IDs
def nextId():
    world["id"] += 1
    return "#" + str(world["id"])

# func to pop the node with the lowest priority from the queue
def pop():
    e = min(q, key=lambda x: x["priority"])
    q.remove(e)
    return e["state"]

isGoal = lambda n: n == goal
children = lambda n: road[n].keys()
actionCost = lambda n, c: road[n][c]

# Initialize dict and priority queue
d[init] = {"parent": None, "cost": 0}  # g(n) = 0 for start
f = lambda n: d[n]["cost"] + heuristics[n]  # A* => f(n) = g(n) + h(n), UCS =>
q.append({"id": nextId(), "state": init, "priority": f(init)})
print("Initial Queue:", q)

# Main loop
while q:
    # Extract the city with the lowest f(n)
    n = pop()
    print("Processing:", n)

    # Check if the goal is reached
    if isGoal(n):
        print("Success")
        break

    # Explore children
    for c in children(n):
        # print(c)
        cost = d[n]["cost"] + actionCost(n, c)  # g(n) + cost of edge
        if c not in d or (cost > 0 and cost < d[c]["cost"]):  # Update if better cost
            d[c] = {"parent": n, "cost": cost}
            q.append({"id": nextId(), "state": c, "priority": f(c)})
        print("Updated PQ= ", q)

# Function to print the path
def path(n):
    if d[n]["parent"] is not None:
        path(d[n]["parent"])
    print(n, end=" -> ")


# Print the path and total cost
print("Final contents of queue")
print(q)
print("Final path to goal")
path(goal)
print("Cost:", d[goal]["cost"])

```

```python
# 4. Write a program that uses graph data of map of Romania to verify if the heuristic value of cities (estimated distance from Bucharest) is consistent.

"""  To verify if the heuristic values of cities in the map of Romania are consistent, we need to check the triangle inequality for the heuristic function h(n). Specifically:

For a heuristic h to be consistent:
h(n) ≤ c(n, c) + h(c)
Where:
  h(n)   : Heuristic value of node n.
  h(c)   : Heuristic value of child c.
  c(n, c): Cost of the edge between n and c.

 """
# Dict of City with name and heuristic value (h(n))
heuristics = {
    "Arad": 366,
    "Bucharest": 0,
    "Fagaras": 176,
    "Oradea": 380,
    "Pitesti": 100,
    "Rimnicu-Vilcea": 193,
    "Sibiu": 253,
}

# Adjacency list => Dict of Dicts (graph with edge costs)
road = {
    "Arad": {"Sibiu": 140},
    "Bucharest": {"Fagaras": 211, "Pitesti": 101},
    "Fagaras": {"Bucharest": 211, "Sibiu": 99},
    "Oradea": {"Sibiu": 151},
    "Pitesti": {"Bucharest": 101, "Rimnicu-Vilcea": 97},
    "Rimnicu-Vilcea": {"Pitesti": 97, "Sibiu": 80},
    "Sibiu": {"Arad": 140, "Fagaras": 99, "Oradea": 151, "Rimnicu-Vilcea": 80},
}

# Function to verify consistency
def is_heuristic_consistent(heuristics, graph):
    for city in graph:
        for neighbor, cost in graph[city].items():
            h_n = heuristics[city]
            h_c = heuristics[neighbor]
            if h_n > cost + h_c:
                print("Inconsistent heuristic detected: h(", city, ") > c(", city, ",", neighbor, ") + h(", neighbor, ")")
                print(h_n, ">", cost, "+", h_c)
                return False
    return True

# Check if the heuristic is consistent
consistent = is_heuristic_consistent(heuristics, road)
if consistent:
    print("The heuristic is consistent for all cities.")
else:
    print("The heuristic is NOT consistent.")
```
