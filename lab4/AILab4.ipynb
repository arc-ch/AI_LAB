```python
#UNIFORM COST SEARCH OR DIJKSTRA'S ALGORITHM 

# MAP OF Romania
city = ["Arad",        # 0
        "Bucharest",   # 1
        "Fagaras",     # 2
        "Oradea",      # 3
        "Pitesti",     # 4
        "Rimnicu-Vilcea",  # 5
        "Sibiu"]       # 6

# Adjacency list: road[x][y] = distance from city x to city y
road = [{6: 140},           # 0 => Arad to Sibiu
        {2: 211, 4: 101},   # 1 => Bucharest to Fagaras, Pitesti
        {1: 211, 6: 99},    # 2 => Fagaras to Bucharest, Sibiu
        {6: 151},           # 3 => Oradea to Sibiu
        {1: 101, 5: 97},    # 4 => Pitesti to Bucharest, Rimnicu-Vilcea
        {4: 97, 6: 80},     # 5 => Rimnicu-Vilcea to Pitesti, Sibiu
        {0: 140, 2: 99, 3: 151, 5: 80}]  # 6 => Sibiu to Arad, Fagaras, Oradea, Rimnicu-Vilcea

# Initialize start (Arad) and goal (Bucharest)
init = 0
goal = 1

# STEP 1: INITIALIZE DICTIONARY AND PRIORITY QUEUE
d = {}  # Stores state details: parent and cost
q = []  # Priority queue to track the next node to explore
world = {'id': -1}  # Unique ID generator for states

# Generate unique IDs for nodes
def nextId():
    world['id'] += 1
    return '#' + str(world['id'])

# Function to extract the node with the minimum priority (lowest cost) from the queue
def pop():
    e = min(q, key=lambda x: x['priority'])  # Find node with minimum cost
    q.remove(e)  # Remove it from the queue
    return e['state']  # Return the state of the node

# Lambda functions for specific operations
isGoal = lambda n: n == goal  # Check if the current state is the goal
children = lambda n: road[n].keys()  # Get all neighboring nodes
actionCost = lambda n, c: road[n][c]  # Get the distance between two cities

# Start by initializing the dictionary and adding the initial state to the queue
d[init] = {'parent': None, 'cost': 0}  # Start state: no parent, cost is 0
f = lambda n: d[n]['cost']  # Priority is based on cost
q.append({'id': nextId(), 'state': init, 'priority': f(init)})  # Add initial state to queue
# print(q)

# STEP 2: WHILE QUEUE IS NOT EMPTY
while q != []:
    # STEP 3: POP THE NODE WITH LOWEST PRIORITY
    n = pop()  # Current state being explored

    # STEP 4: CHECK IF GOAL STATE IS REACHED
    if isGoal(n):
        print("Success! Goal reached.")
        break

    # STEP 5: FOR EACH CHILD OF THE CURRENT NODE
    for c in children(n):
        print(c)
        # Calculate cost to reach child c from the current node n
        cost = d[n]['cost'] + actionCost(n, c)

        # Update dictionary and queue if:
        # 1. Child c is not already visited OR
        # 2. A lower cost path to child c is found
        if c not in d or (cost > 0 and cost < d[c]['cost']):
            d[c] = {'parent': n, 'cost': cost}  # Update dictionary
            q.append({'id': nextId(), 'state': c, 'priority': f(c)})  # Add child to queue
        print(q)
# STEP 6: TRACE AND PRINT THE PATH
def path(n):
    if d[n]['parent'] is not None:
        path(d[n]['parent'])  # Recursive call to parent
    print(city[n], end=" -> ")

# Print the queue, path and total cost
print("Final contents of queue")
print(q)
print("Final path to goal")
path(goal)
print(f"Total Cost: {d[goal]['cost']}")
print(type(road))
```

```python
#HEURISTIC -> A* ALGORITHM
# 1) adding heuristic value in city
# 2) changing priority function to g(n) + h(n)


# Dict of City with name and heuristic value
city = {
    "Arad": 366,
    "Bucharest": 0,
    "Fagaras": 176,
    "Oradea": 380,
    "Pitesti": 100,
    "Rimnicu-Vilcea": 193,
    "Sibiu": 253,
}

# Adjacency list => Dict of Dicts
road = {
    "Arad": {"Sibiu": 140},                     # Arad to Sibiu=> distance = 140
    "Bucharest": {"Fagaras": 211, "Pitesti": 101},
    "Fagaras": {"Bucharest": 211, "Sibiu": 99},
    "Oradea": {"Sibiu": 151},
    "Pitesti": {"Bucharest": 101, "Rimnicu-Vilcea": 97},
    "Rimnicu-Vilcea": {"Pitesti": 97, "Sibiu": 80},
    "Sibiu": {"Arad": 140, "Fagaras": 99, "Oradea": 151, "Rimnicu-Vilcea": 80},
}

# Start and goal
init = "Arad"  # Starting city
goal = "Bucharest"  # Goal city

# Priority queue and tracking
d = {}
q = []
world = {"id": -1}

# Function to return unique IDs
def nextId():
    world["id"] += 1
    return "#" + str(world["id"])

# Function to pop the node with the lowest priority from the queue
def pop():
    e = min(q, key=lambda x: x["priority"])  # Find element with the smallest f(n)
    q.remove(e)
    return e["state"]

isGoal = lambda n: n == goal
children = lambda n: road[n].keys()
actionCost = lambda n, c: road[n][c]

# Initialize dictionary and priority queue
d[init] = {"parent": None, "cost": 0}  # g(n) = 0 for start
f = lambda n: d[n]["cost"] + city[n]                                         # f(n) = g(n) + h(n)
q.append({"id": nextId(), "state": init, "priority": f(init)})
print("Initial Queue:", q)

# Main loop
while q:
    # Extract the city with the lowest f(n)
    n = pop()
    print("Processing:", n)

    # Check if the goal is reached
    if isGoal(n):
        print("Success")
        break

    # Explore children
    for c in children(n):
        # print(c)
        cost = d[n]["cost"] + actionCost(n, c)  # g(n) + cost of edge
        if c not in d or (cost > 0 and cost < d[c]["cost"]):  # Update if better cost
            d[c] = {"parent": n, "cost": cost}
            q.append({"id": nextId(), "state": c, "priority": f(c)})
        print("Updated PQ= ", q)

# Function to print the path
def path(n):
    if d[n]["parent"] is not None:
        path(d[n]["parent"])
    print(n, end=" -> ")

# Print the path and total cost
print("Final contents of queue")
print(q)
print("Final path to goal")
path(goal)
print("Cost:", d[goal]["cost"])

```
